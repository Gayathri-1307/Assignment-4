```python
import seaborn as sns
import pandas as pd
heart=pd.read_csv("heart_disease.csv")
heart.head()
```

# Relation plots in seaborn

```python
sns.set_style("darkgrid")
sns.relplot(x="Age",
            y="RestingBP",
            hue="Gender",
            data=heart)
```

# Scatter plot

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Age",
            y="RestingBP",
            hue="Gender",
            marker="p",
            data=heart)
```

# Line plot

```python
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
sns.lineplot(x="Age",
             y="HeartDisease",
             hue="Gender",
             data=heart)
plt.show()
```

# Categorical plots
Barplot

```python
sns.barplot(x="Gender",
             y="HeartDisease",
             palette="plasma",
             data=heart)
```

# Count plot

```python
sns.countplot(x="HeartDisease",
             palette="plasma",
             data=heart)
```

# Box plot

```python
sns.boxplot(x="HeartDisease",
             y="Age",
             hue="Gender",
             data=heart)
```

# ViolinPlot

```python
sns.violinplot(x="HeartDisease",
            y="Age",
            hue="Gender",
            split=True,
             data=heart)
```

# Stripp plot

```python
sns.stripplot(x="HeartDisease",
              y="Age",
              hue="Gender",
              jitter=True,
              dodge=True,
              data=heart)
```

# Swarmplot

```python
sns.swarmplot(x="HeartDisease",
             y="Age",
             hue="Gender",
             data=heart)
```

```python
sns.violinplot(x="HeartDisease",
            y="Age",
            hue="Gender",
            split=True,
             data=heart)
sns.swarmplot(x="HeartDisease",
             y="Age",
             hue="Gender",
             color="black",
             data=heart)
```

# Factor plot

```python
import seaborn as sns
sns.factorplot(x="HeartDisease",
             y="Age",
             hue="Gender",
             kind="bar",
             data=heart)
```

# Distribution plot

# Histogram plot

```python
import seaborn as sns
import pandas as pd
heart=pd.read_csv("heart_disease.csv")
sns.histplot(x="Age",
             y="ST_Slope",
             hue="Gender",
             data=heart)
```

# Distribution plot


```python
sns.distplot(heart["Age"],kde=True,color="red",bins=30)
```

# Joint plot


```python
sns.jointplot(x="Age",
             y="HeartDisease",
             hue="Gender",
             data=heart)
```

# Pair plot

```python
sns.pairplot(data=heart,hue="Gender")
```

# Rug plot

```python
sns.rugplot(data=heart,x="Age")
```

# kdeplot

```python
sns.kdeplot(data=heart,x="Age")
```

# Regression plots

# lmplot

```python
sns.lmplot(x="Cholesterol",
           y="RestingBP",
           data=heart,
          hue="Gender",
          markers=["o","v"],
          scatter_kws={"s":100},
          palette="plasma")
```

# regplot

```python
sns.regplot(x="Cholesterol",
           y="RestingBP",
           data=heart,
          scatter_kws={"s":100})
```

# Heatmap
